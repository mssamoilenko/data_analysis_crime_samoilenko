```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn import tree
```

```python
data ='/kaggle/input/crime-cast-forecasting-crime-categories/train.csv'

df = pd.read_csv(data)
df.head()
```

```python
df.describe(include='all')
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns = df.columns.str.lower()
df.head()
```

```python
filtered_crimes = df[
    (df['Crime_Category'] == 'Property Crimes') & 
    (df['Area_Name'] == '77th Street') &
    (df['Victim_Age'] > 18)
]
```

```python
agg_results = {
    'Total Crimes': df['Crime_Category'].count(),
    'Max Age': df['Victim_Age'].max(),
    'Min Age': df['Victim_Age'].min(),
    'Average Age': round(df['Victim_Age'].mean(), 1),
    'Most Common Weapon': df['Weapon_Description'].mode()[0],
    'Crimes at Streets': df[df['Premise_Description'] == 'STREET']['Premise_Description'].count()
}
```

```python
sorted_df = df.sort_values(['Victim_Age', 'Time_Occurred'], ascending=[False, True])
```

```python
grouped_data = df.groupby(['Crime_Category', 'Victim_Sex']).agg({
    'Victim_Age': ['mean', 'count'],
    'Weapon_Used_Code': 'nunique'
}).reset_index()
grouped_data.columns = ['Crime Category', 'Victim Sex', 'Average Age', 'Crime Count', 'Unique Weapons']
```

```python
pivot_table = pd.pivot_table(
    df,
    values='Victim_Age',
    index='Crime_Category',
    columns='Area_Name',
    aggfunc=np.mean,
    fill_value=0
)
```

```python
print("Фільтровані дані (Property Crimes у 77th Street для жертв старше 18):")
print(filtered_crimes[['Location', 'Victim_Age', 'Victim_Sex', 'Weapon_Description']].head())
```

```python
print("\nАгреговані результати:")
for key, value in agg_results.items():
    print(f"{key}: {value}")
```

```python
print("\nПерші 5 записів після сортування (вік та час злочину):")
print(sorted_df[['Victim_Age', 'Time_Occurred', 'Crime_Category']].head())
```

```python
print("\nГрупування за категорією злочину та статтю жертви:")
print(grouped_data)
```

```python
print("\nЗведена таблиця (середній вік жертв за категорією та районом):")
print(pivot_table)
```
